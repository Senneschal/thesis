```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
def deform(array, factor, axis):
    stretched = np.repeat(array, np.ceil(factor), axis=axis)
    indices = np.linspace(0, stretched.shape[axis] - 1, int(array.shape[axis] * factor))
    indices = np.rint(indices).astype(int)
    return np.take(stretched, indices, axis=axis)
```

```python
for i in range(733, 1018):
    if(os.path.exists('D:\Lungs\_'+str(i)+'.npz')):
        
        lung = np.load('D:\Lungs\_'+str(i)+'.npz')['arr_0']
        nods = np.load('D:\Cropped\_'+str(i)+'.npz')['arr_0']
        
        nonz = np.nonzero(lung)
        
        lung = lung[nonz[0].min():nonz[0].max()+1, nonz[1].min():nonz[1].max()+1, nonz[2].min()+5:nonz[2].max()-5]
        nods = nods[nonz[0].min():nonz[0].max()+1, nonz[1].min():nonz[1].max()+1, nonz[2].min()+5:nonz[2].max()-5]
        
        lung = deform(lung, 256/lung.shape[0], axis=0)
        lung = deform(lung, 256/lung.shape[1], axis=1)
        lung = deform(lung, 64/lung.shape[2], axis=2)

        nods = deform(nods, 256/nods.shape[0], axis=0)
        nods = deform(nods, 256/nods.shape[1], axis=1)
        nods = deform(nods, 64/nods.shape[2], axis=2)
        
        np.savez_compressed('l_'+str(i)+'.npz', lung)
        np.savez_compressed('n_'+str(i)+'.npz', nods)
        
        print(i)
```
