```python
import pylidc as pl
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
volumes = []

for i in range(10000):
    print(i)
    volumes.append(pl.query(pl.Annotation)[i].volume)
```

```python
hist = np.histogram(volumes, bins=5000)
```

```python
hist
```

```python
plt.plot(hist[1][0:300], hist[0][0:300])
plt.plot(np.arange(10000), 150000*(np.arange(10000)**(-1.5)),'--')
plt.xlabel('Nodule volume [mm³]')
plt.ylabel('Number of nodules')
plt.xscale('log')
plt.yscale('log')
plt.xlim(5,2200)
plt.ylim(0.5,500)
plt.savefig('plotr')
```
